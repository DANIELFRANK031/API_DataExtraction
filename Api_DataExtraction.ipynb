```python
##importing Api file from our file source
import requests
import pandas as pd
```

```python
## calling for my api request
call=requests.get("https://v6.vbb.transport.rest/locations?query=berlin")
```

```python
##checking for request status. 200= successful
call.status_code
```

```python
##Exploring my Api file to show results of call
call.json()
```

```python
##Exploring my data list  
call.json()[0]
```

```python
##checking for dictionary keys from call
call.json()[0].keys()
```

```python
call.json()
```

```python
##Assigning my result to a new value X 
X=call.json()
```

```python
##importing results into my dataframe for more analysis
df=pd.DataFrame.from_records(X)
```

```python
##Exploring five 5 information from the data set
df.head()
```

```python
##Extracting needed colunm for analysis then assigning the data to a new variable (newdf)
newdf=df[['type','id','name','stationDHID']]
```

```python
newdf.head()
```

```python
##Extracting only the name to an empty list
name=[]
```

```python
##Using the FOR loop to extract information into the empty list
for x in call.json():
    name.append(x['name'])
##print(name)
df_name=pd.DataFrame(name)##converting to a DataFrame
```

```python
df_name
```

```python
##Using the FOR loop to extract information into the empty list
longitude=[]
for x in call.json():
    longitude.append(x['location']['longitude'])
print(longitude)
```

```python
##Using the FOR loop to extract information into the empty list
latitude=[]
for x in call.json():
    latitude.append(x['location']['latitude'])
print(latitude)
```

```python
##Using the FOR loop to extract information into the empty list
products=[]
for x in call.json():
    products.append(x['products'])
#print(products)
df_products=pd.DataFrame(products)##converting to a DataFrame
```
